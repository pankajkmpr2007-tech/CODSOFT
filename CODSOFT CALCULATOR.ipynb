```python
import tkinter as tk
from tkinter import messagebox

# --- Colors & Aesthetics ---
# Modern Dark Palette
COLOR_BG = "#121212"          # Very dark background
COLOR_DISPLAY_BG = "#1E1E1E" # Slightly lighter display background
COLOR_TEXT = "#FFFFFF"        # White text
COLOR_BTN_DEFAULT = "#333333" # Dark gray buttons (numbers)
COLOR_BTN_HOVER = "#424242"   # Lighter gray when hovering
COLOR_BTN_ACCENT = "#FF9F43"  # Orange for operators (+, -, *, /)
COLOR_BTN_ACCENT_HOVER = "#FFB366"
COLOR_BTN_EQUAL = "#2ecc71"   # Green for equal button
COLOR_BTN_EQUAL_HOVER = "#4cd137"
COLOR_BTN_CLEAR = "#e74c3c"   # Red for clear button
COLOR_BTN_CLEAR_HOVER = "#ff6b6b"

# Font configurations
FONT_DISPLAY = ("Consolas", 36, "bold") # Digital look
FONT_BUTTON = ("Arial", 16)
FONT_SPECIAL = ("Arial", 16, "bold")

# --- Logic Functions ---
def press_key(char):
    current = display_var.get()
    # Prevent leading zeros unless it's just '0.'
    if current == "0" and char != ".":
        display_var.set(str(char))
    else:
        display_var.set(current + str(char))

def clear_display():
    display_var.set("")

def backspace():
    current = display_var.get()
    display_var.set(current[:-1])

def calculate_result():
    try:
        expression = display_var.get()
        # Replace mathematical symbols for evaluation
        expression = expression.replace('×', '*').replace('÷', '/')
        
        # Safe evaluation
        # Note: In production, use a parser instead of eval for better security
        result = eval(expression)
        
        # Formatting result: remove .0 if it's an integer
        if isinstance(result, float) and result.is_integer():
            result = int(result)
            
        display_var.set(result)
    except ZeroDivisionError:
        messagebox.showerror("Error", "Division by zero is undefined.")
        display_var.set("Error")
    except Exception:
        messagebox.showerror("Error", "Invalid Expression")
        display_var.set("Error")

# --- Interactive Styling Functions ---
def on_enter(e, default_color):
    """Changes button background on mouse hover."""
    e.widget['background'] = default_color

def on_leave(e, default_color):
    """Restores button background when mouse leaves."""
    e.widget['background'] = default_color

def create_button(parent, text, row, col, command, btn_color=COLOR_BTN_DEFAULT, hover_color=COLOR_BTN_HOVER, font=FONT_BUTTON, columnspan=1):
    """Helper function to create styled, interactive buttons."""
    btn = tk.Button(
        parent, 
        text=text, 
        font=font, 
        bg=btn_color, 
        fg=COLOR_TEXT, 
        bd=0, # No border for flat design
        highlightthickness=0,
        activebackground=hover_color, # Color when clicked
        activeforeground=COLOR_TEXT,
        command=command,
        # Padding inside the button text
        pady=15 
    )
    # Using grid with padding between buttons
    btn.grid(row=row, column=col, columnspan=columnspan, padx=3, pady=3, sticky="nsew")
    
    # Bind hover events
    btn.bind("<Enter>", lambda e: on_enter(e, hover_color))
    btn.bind("<Leave>", lambda e: on_leave(e, btn_color))
    
    return btn

# --- Main App Setup ---
root = tk.Tk()
root.title("Modern Calculator")
root.geometry("340x510") # Sleeker aspect ratio
root.configure(bg=COLOR_BG)

# Optional: Add an icon (if you have one, e.g., 'icon.ico')
# root.iconbitmap('icon.ico') 

display_var = tk.StringVar()

# 1. --- Display Area ---
display_frame = tk.Frame(root, bg=COLOR_BG, pady=10)
display_frame.pack(side=tk.TOP, fill=tk.BOTH)

display = tk.Entry(
    display_frame, 
    textvariable=display_var, 
    font=FONT_DISPLAY, 
    bg=COLOR_DISPLAY_BG, 
    fg=COLOR_TEXT, 
    borderwidth=0, 
    justify="right",
    insertbackground=COLOR_TEXT, # Cursor color
)
display.pack(side=tk.TOP, fill=tk.BOTH, padx=15, pady=15, ipady=15) # ipady adds internal vertical spacing

# 2. --- Buttons Area (Grid) ---
buttons_frame = tk.Frame(root, bg=COLOR_BG)
buttons_frame.pack(side=tk.TOP, fill=tk.BOTH, expand=True, padx=10, pady=10)

# Configure rows and columns weights for dynamic resizing
for i in range(6):
    buttons_frame.rowconfigure(i, weight=1)
for j in range(4):
    buttons_frame.columnconfigure(j, weight=1)

# --- Define and Place Buttons ---

# Row 1: Utility functions
create_button(buttons_frame, "C", 0, 0, clear_display, btn_color=COLOR_BTN_CLEAR, hover_color=COLOR_BTN_CLEAR_HOVER, font=FONT_SPECIAL, columnspan=2)
create_button(buttons_frame, "⌫", 0, 2, backspace, font=FONT_SPECIAL)
create_button(buttons_frame, "÷", 0, 3, lambda: press_key('/'), btn_color=COLOR_BTN_ACCENT, hover_color=COLOR_BTN_ACCENT_HOVER, font=FONT_SPECIAL)

# Row 2: Numbers 7-9 & Multiply
create_button(buttons_frame, "7", 1, 0, lambda: press_key(7))
create_button(buttons_frame, "8", 1, 1, lambda: press_key(8))
create_button(buttons_frame, "9", 1, 2, lambda: press_key(9))
create_button(buttons_frame, "×", 1, 3, lambda: press_key('×'), btn_color=COLOR_BTN_ACCENT, hover_color=COLOR_BTN_ACCENT_HOVER, font=FONT_SPECIAL)

# Row 3: Numbers 4-6 & Subtract
create_button(buttons_frame, "4", 2, 0, lambda: press_key(4))
create_button(buttons_frame, "5", 2, 1, lambda: press_key(5))
create_button(buttons_frame, "6", 2, 2, lambda: press_key(6))
create_button(buttons_frame, "−", 2, 3, lambda: press_key('-'), btn_color=COLOR_BTN_ACCENT, hover_color=COLOR_BTN_ACCENT_HOVER, font=FONT_SPECIAL)

# Row 4: Numbers 1-3 & Add
create_button(buttons_frame, "1", 3, 0, lambda: press_key(1))
create_button(buttons_frame, "2", 3, 1, lambda: press_key(2))
create_button(buttons_frame, "3", 3, 2, lambda: press_key(3))
create_button(buttons_frame, "+", 3, 3, lambda: press_key('+'), btn_color=COLOR_BTN_ACCENT, hover_color=COLOR_BTN_ACCENT_HOVER, font=FONT_SPECIAL)

# Row 5: 0, Decimal, Equal
create_button(buttons_frame, "0", 4, 0, lambda: press_key(0), columnspan=2)
create_button(buttons_frame, ".", 4, 2, lambda: press_key('.'))
create_button(buttons_frame, "=", 4, 3, calculate_result, btn_color=COLOR_BTN_EQUAL, hover_color=COLOR_BTN_EQUAL_HOVER, font=FONT_SPECIAL)

# Start the application loop
root.mainloop()
```
